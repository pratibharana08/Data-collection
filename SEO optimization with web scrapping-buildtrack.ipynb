We conclude that the title has on average 11 words with around 71 characters. Further some advancements can be made to analyse how title affect more clicks.

```python
#Lets dig further to find the most frequently used words in title
title_words=str(buildtrack_data['title'].tolist()).replace('_', ' ').replace('|',' ').replace('-',' ')
adv.word_frequency(title_words, rm_words=adv.stopwords['english']).head(20)
```

```python
#most frequesnt used 2 words and we can analyze different lenght using phrase_len
adv.word_frequency(title_words, rm_words=adv.stopwords['english'], phrase_len=2).head(20)
```

```python
#lets create a wordcloud now
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(title_words)
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()
```

We can now clearly see the most used word and can hence modify the marketing strategy using the information.
